# Current World Population

```python
# Library
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Sending Request
page_url = "https://www.worldometers.info/world-population/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'
}

page = requests.get(page_url,
                    headers = headers)
print(page)

# Beautiful Soup Object
soup = BeautifulSoup(page.content,'html.parser')
soup

# Our Table Current World Population
table2 = soup.find('table',
                   {'class':'table table-striped table-bordered table-hover table-condensed table-list'})

table2

# Table Headers Current World Population
headers = []
for i in table2.find_all('th'):
    col_name = i.text
    headers.append(col_name)
    
# Data Frame Columns Name
df = pd.DataFrame(columns=headers)
df

# Table Data Current World Population
for i in table2.find_all('tr')[1:]:
    row_data = i.find_all('td')
    rows = [t.text for t in row_data]
    lenght = len(df)
    df.loc[lenght] = rows
df

# The End..!
```
